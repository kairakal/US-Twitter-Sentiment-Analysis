```python

from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

    
```

```python
from nltk import pos_tag
```

```python
from nltk.corpus import wordnet

def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops=set(stopwords.words('english'))
punctuation=list(string.punctuation)
stops.update(punctuation)
stops,string.punctuation
```

```python
# removing stop words and lemmatizer
from time import time
def clean_review(words):
    global count
    s=time()
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag(w)
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word)
    e=time()
    count+=1
    print("doc no = ",count, "time = ",e-s)
    return   output_words          
```

```python
from nltk import pos_tag
count=0
documents=[(clean_review(document),category)for document,category in documents]
```

```python
count=0
documents_test=[(clean_review(document))for document in documents_test]
```

```python
all_words=[]
for doc in documents:
    all_words+= doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features=[i[0]  for i in common]
```

```python
def get_feature_dict(words):
    current_feature={}
    words_set=set(words)
    for w in features:
        current_feature[w]=w in words_set
    return current_feature    
```

```python
output=get_feature_dict(documents[0][0])
output
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in documents]
```

```python
testing_data=[get_feature_dict(doc) for doc in documents_test]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
y_pred = []
for doc in testing_data:
    y_pred.append(classifier.classify(doc))
```
